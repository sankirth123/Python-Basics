# 3 Describe three different data types
Python has different data types,Variables can store data of different types and different types can do different things.
The three data types are

1.Numeric type : Int, Float, Complex

2.Text type : Str

3.Sequence type : List, Tuple, Set

1. Numeric type

Int : Int or integer, which has positive or negative number without decimals is called integer
Float : Float or floating point number is a number, which has positive or negative sign containing one or more decimals is called float 
Complex : Complex numbers are written with a "j" as the imaginary part

Below are the examples of Int, Float, Complex 

Example:
x = 5,

y = 2.5,

z = 9j.


print(type(x)),

print(type(y)),

print(type(z)),

Output : int, float, complex


2. Text type : Str

str : strings in python are surrounded by either single quotation marks or double quotation marks.

Example:
I = "iNeuron" 

print(type(I))


O/P : Str

3. Sequence type : List, Tuple, Set

List : List are used to store multiple items and created using square brackets

Example:
B = ["Apple","Ball","Car" ]


print(B)

O/P : ["Apple","Ball","Car" ]


Tuple : Tuples are used to store multiple items in a single variable. It is a collection which is ordered and unchangeable, tuple  written with round brackets.

Example:
C = ("Apple", "ball", "car")

print(C)

O/P : ('Apple', 'ball', 'car')


Set : Sets are used to store multiple items in a single variable. It is a collection which is both unordered and unindexed,
Sets are written with curly brackets.

Example:
d = {"Apple", "ball", "car"}

print(d)

O/P : {'ball', 'Apple', 'car'}
